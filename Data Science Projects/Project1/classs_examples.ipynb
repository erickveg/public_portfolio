```python
q1 = names.query('name == "Stewart"')

chart2 = (alt.Chart(q1).encode(
    x="year",
    y="Total",
    )
    .mark_line())

chart2
```

```python
q1_dot = q1.query('year == 1972')
q1_dot
```

Create a point in the chart 

```python
chart1 = (alt.Chart(q1_dot).mark_circle(color="red").encode(x = "year", y = 'Total'))
```

Create a line

```python
chart3 = (alt.Chart(q1_dot).mark_rule(color="red").encode(x = "year"))
```
